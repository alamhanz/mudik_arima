```python
def status(city):
    jabo=['jakarta','bogor','depok','tangerang','bekasi']
    r=0
    for j in jabo:
        if j in city.lower():
            r=1
    return r
```

```python
status('solo')
```

```python
df3['is_jabodetabek']=df3.lokasi_dominan.apply(status)
```
